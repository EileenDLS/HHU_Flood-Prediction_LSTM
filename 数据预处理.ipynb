```python
def time_chg(data, L):
    for i in tqdm(range(L)):
        t = data.iloc[i]
        
        y = str(t)[:4]
        m = str(t)[4:6]
        d = str(t)[6:8]
        h = str(t)[8:]

        t_ = y+'-'+m+'-'+d+' '+h
        t__ = pd.Timestamp(t_)
        data.iloc[i] = t__
    return data
```

```python
def Add_row(A, B):
    a = np.array(A)
    b = np.array(B)
    c = np.concatenate((b,a), axis=0)
    
    return c
```
